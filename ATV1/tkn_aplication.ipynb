```python
import tkn_functions as tknF
import json
import time
```

```python
# Abre o arquivo JSON e carrega o conteúdo
with open("corpus_completo.json", "r", encoding="utf-8") as arquivo:
    dados = json.load(arquivo)

# pega o texto do JSON
texto = dados["text"]
```

```python
# treinando o tokenizador e retornando um script de merges associando um par a um novo id, 
# uma lista de adjecencia de possiveis merges para cada id unico e
# um vocabulario de bytes para decodificação

# recebe um texto e o hyperparametro de vocabularios desejado
merge_script, id_adj_list, vocab = tknF.train(texto, 500)
```

```python
merge_script
```

```python
# a lista de adjecencia pode ter muitas chaves vazias, dessa forma pego as que possuem conteudo
clean_adj_list = {}
for i in range(len(id_adj_list)):
    if id_adj_list[i] != []:
        clean_adj_list[i] = id_adj_list[i]
   
clean_adj_list
```

```python
# Abre o arquivo JSON e carrega o conteúdo
with open("248.json", "r", encoding="utf-8") as arquivo:
    dados_teste = json.load(arquivo)

# pega o texto do JSON
texto_teste = dados_teste["text"]
```

```python
encoded_txt = tknF.encode(texto_teste, merge_script, id_adj_list)
encoded_txt
```

```python
decoded_ids = tknF.decode(encoded_txt, vocab)
decoded_ids
```
